```python
cols_drop=["Pool QC","Misc Feature","Alley","Fence","Order","PID"]
df=df.drop(columns=cols_drop)
```

```python
df.info()
```

```python
# filling the missing values
# in categorical data we fill be filling none
cat_fill_none=["Mas Vnr Type",
    "Fireplace Qu",
    "Garage Type",
    "Garage Finish",
    "Garage Qual",
    "Garage Cond",
    "Bsmt Qual",
    "Bsmt Cond",
    "Bsmt Exposure",
    "BsmtFin Type 1",
    "BsmtFin Type 2"]
for col in cat_fill_none:
    if col in df.columns:
      df.fillna({col:"None"},inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
#numerical data fill median value
num_fill_median=["Lot Frontage","Garage Yr Blt","Mas Vnr Area"]
for col in num_fill_median:
    if col in df.columns:
      df.fillna({col:df[col].median()},inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
Zero_fill=["Bsmt Full Bath","Bsmt Half Bath","Bsmt Unf SF"]
for col in Zero_fill:
    if col in df.columns:
      df.fillna({col:0},inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
#Electrical is object type data
if "Electrical" in df.columns:
  df["Electrical"]=df["Electrical"].fillna(df["Electrical"].mode()[0])
```

```python
df.isnull().sum().sort_values(ascending=False).head()
```

```python
Zero_fill=["Garage Cars","BsmtFin SF 2","Total Bsmt SF","BsmtFin SF 1","Garage Area"]
for col in Zero_fill:
    if col in df.columns:
      df.fillna({col:0},inplace=True)
```
